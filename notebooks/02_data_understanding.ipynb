Data Understanding
orders-PK-orderID , FK-customerID
customers-PK-customerID , CK - customeruniqueID
payments-FK-orderID
reviews- PK-reviewID , FK-orderID

The orders table acts as the central fact table, linked to customers, payments, and reviews via foreign keys. For churn analysis, I use customer_unique_id to track real customers across multiple orders
